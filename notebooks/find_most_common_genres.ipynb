```python
import os
import pandas as pd
```

## Load the data 

```python
data=pd.read_csv('../data_local/tracks_with_genres.csv')
##data.info()
```

```python
#genres_list = data.genres.unique().tolist()
genres_list = data.genres.tolist()
#genres_list
```

## Prepare the genre list to perform map reduce

```python
import re

modified_list = []
regex = re.compile('[^a-zA-Z, ]')

def fix_the_list(l):
    for i in range(len(l)):
            l[i]=l[i].strip()
            word=regex.sub("", l[i])
            word = word.split(',')
            for i in range(len(word)):
                word[i] = word[i].lstrip()
                if word[i] != "":
                    modified_list.append(word[i])
                    
fix_the_list(genres_list)
```

```python
from contextlib import redirect_stdout

with open('out.txt', 'w') as f:
    with redirect_stdout(f):
        for el in modified_list:
            print ('%s\t%s' % (el, 1))
```

## Run the reducer through the terminal and check the results

```python
cat out.txt |sort -k1,1| python reduce.py  > output_reducer.txt
```

## Check the contents 

```python
f = open('output_reducer.txt', 'r')
file_contents = f.read()
print (file_contents)
f.close()
```

```python
top_genres = file_contents.split('\n')
top_glist = []

for line in top_genres:
    line.strip(" ")
    top_glist.append(line.split("\t"))
    
#columns_df = ['Genre', 'Count']
genres_df = pd.DataFrame(data = top_glist, columns =['Genre', 'Count'] )
```

## Display top 15 genres by count

```python
genres_df['Count'] = pd.to_numeric(genres_df['Count'])
df_top15 = genres_df.nlargest(15, 'Count')
top_15_list = df_top15['Genre'].tolist()
```
